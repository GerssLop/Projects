# Prueba de hipótesis

Mostrando info de SQL result 07

```python
print('project_sql_result_07')
df_sql_07.info()
```

Transformación de un tipo String a un tipo Time

```python
df_sql_07['start_ts'] = pandas.to_datetime(df_sql_07['start_ts'],format='%Y.%m.%d %H:%M:%S')

df_sql_07.info()
print()
print('Los valores duplicados en este DataFrame son',df_sql_07.duplicated().sum())
df_sql_07.describe()
```

Mostrando los valores

```python
print(df_sql_07[df_sql_07.duplicated()])
```

```python
df_sql_07=df_sql_07.drop_duplicates()
print(df_sql_07[df_sql_07.duplicated()])
print('Los valores duplicados son:',df_sql_07.duplicated().sum())
```

Los valores que contienen es acerca de los viajes desde Loop hasta el A. I. O'Hare; del cual se 
establecen dos condiciones "Bad" y "Good"

```python
for i in range(len(df_sql_07)):
    if(df_sql_07.iloc[i]['start_ts'].weekday()!=5):
        print('No es el día')
```

Separando por condición

```python
alpha = 0.05

condition_good=df_sql_07[df_sql_07['weather_conditions']=='Good']
condition_bad=df_sql_07[df_sql_07['weather_conditions']=='Bad']
levene_stat, p_levene = st.levene(condition_good['duration_seconds'], condition_bad['duration_seconds'])

print(f'Valor p de la prueba de Levene: {p_levene:.12f}')

equal_var = p_levene > alpha

results = st.ttest_ind(condition_good['duration_seconds'], condition_bad['duration_seconds'], equal_var=equal_var)

p_value_rounded = round(results.pvalue, 12)
print('Valor p de la prueba t:', p_value_rounded)

if (results.pvalue<alpha):
    print("Diferencia significativa")
else:
    print("No hay diferencia")
```

 <div class="alert alert-block alert-danger">
<b>Comentario revisor</b> <a class="tocSkip"></a>

Para esta prueba te recomiendo hacer una prueba de  Levene  para mostrar si las varianzas son iguales y agregarlo dentro de la función. Actualmente lo colocas como equal_var=False. Para esto, primero debes calcular las varianzas para cada uno de los días (lluvioso y no lluvioso) debes de hacer uso de la siguiente función:
    
    levene(condition_good['duration_seconds'],condition_bad['duration_seconds'])
    
Adoptaremos un nivel de significancia de 0.05; si el valor p resultante es mayor a 0.05, no podemos rechazar la hipótesis nula, y si es menor a 0.05, rechazamos la hipótesis nula, indicando que las varianzas no son iguales.
    
Solamente recuerda que la prueba de levene no es sustituto a la prueba de st.ttest_ind, más bien es complemento para saber que colocar dentro del elemento "equal_var".  Es por eso que para terminar la prueba debes de realizar la prueba de st.ttest_ind considernado el resultado de la prueba realizada de levene.
